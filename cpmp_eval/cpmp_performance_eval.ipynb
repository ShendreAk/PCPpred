```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import pearsonr, spearmanr
```

```python
df_pampa = pd.read_csv('Test.csv')
df_pampa
```

```python
df = df_pampa['SMILES']
df
```

```python
df.to_csv('pampa_smiles.csv', index=False)
```

```python
y_test = df_pampa['Permeability']
```

```python
df_predict = pd.read_csv('pampa_cpmp_predictions.csv')
predictions = df_predict['pampa']
```

```python
mse_test = mean_squared_error(y_test, predictions)
print(mse_test)
mae_test = mean_absolute_error(y_test, predictions)
print(mae_test)
rmse_test = np.sqrt(mse_test)
print(rmse_test)
r2_test = r2_score(y_test, predictions)
print(r2_test)
pearson_test, _ = pearsonr(y_test, predictions)
print(pearson_test)
spearman_test, _ = spearmanr(y_test, predictions)
print(spearman_test)
```
